```python
import pandas as pd
import matplotlib.pyplot as plt

# Load metrics
df = pd.read_csv("metrics.csv")

# Separate out the rows for each metric
train_df = df.dropna(subset=["train_acc_epoch"])[["epoch", "train_acc_epoch"]]
val_df = df.dropna(subset=["val_acc"])[["epoch", "val_acc"]]

# Merge on epoch
merged = pd.merge(train_df, val_df, on="epoch")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(merged["epoch"], merged["train_acc_epoch"], label="Train Accuracy")
plt.plot(merged["epoch"], merged["val_acc"], label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Train vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.savefig("accuracy_plot.png")
print("✅ Saved plot as accuracy_plot.png")
```
