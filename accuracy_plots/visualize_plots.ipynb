```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load metrics
df = pd.read_csv("distilebert_metrics.csv")

# Separate out the rows for each metric
train_df = df.dropna(subset=["train_acc_epoch"])[["epoch", "train_acc_epoch"]]
val_df = df.dropna(subset=["val_acc"])[["epoch", "val_acc"]]

# Merge on epoch
merged = pd.merge(train_df, val_df, on="epoch")

# Plot
plt.figure(figsize=(8, 5))  # Initialize a new plot
plt.plot(merged["epoch"], merged["train_acc_epoch"], label="Train Accuracy")    # Plot train accuracy vs epochs
plt.plot(merged["epoch"], merged["val_acc"], label="Validation Accuracy")   # Plot validation accuracy vs epochs
plt.xlabel("Epoch") # Label x-axis as Epoch
plt.ylabel("Accuracy")  # Label y-axis as Accuracy
plt.title("Train vs Validation Accuracy")   # Title of the plot
plt.legend()    # Add legend to differentiate train and validation curves
plt.grid(True)  # Add grid for better visualization
plt.savefig("accuracy_plot.png")    # Save the plot to a file
print("Saved plot as accuracy_plot.png")
```
